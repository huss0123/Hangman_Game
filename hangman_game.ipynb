```python
import random as r 
```

```python
#functions 

def get_index( word:str , letter:str , start:int ):        
    return word.index(letter,start+1)
```

```python
# list of fruits to choose from
list =["apple", "banana", "grape", "orange", "mango", "kiwi", "strawberry", "watermelon", "peach", "cherry"]
# variables
word = r.choice(list)
word_length = len(word)
display = []
chances = word_length + 2

for i in range(word_length):
    display.append("-")
print("Welcome to HANGMAN GAME\nTry to guess the hidden fruit")
print(f"the fruit contains {word_length} letters")
answer = ""
for d in display:
    answer += d
print (f"'{answer}'")
```

```python



for i in range(chances) :
    
    
    user_input= input("enter a valid letter").lower().strip()

    # check validation
    if not user_input.isalpha():
        print("enter a valid letter")
        continue
    
    # check correctness
    if user_input in word:
        print(f"{user_input} is correct ")
        index = word.index(user_input)
        display[index] = user_input
        
        
        pass
    else:
        print(f"{user_input} is wrong , try another letter")
        pass
    answer = ""
    for d in display:
        answer += d
    print (answer)
    print(f"you still have {chances-i-1} chances")
```
