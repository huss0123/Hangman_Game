```python
# list of fruits to choose from
list =["apple", "banana", "grape", "orange", "mango", "kiwi", "strawberry", "watermelon", "peach", "cherry"]
# variables
#word = r.choice(list)                  #uncomment
word = "grape"                         #comment
word_length = len(word)
display = ["-"]*word_length
chances = word_length + 2


print("Welcome to HANGMAN GAME\nTry to guess the hidden fruit")
print(f"the fruit contains {word_length} letters")

answer = "".join(display)

print (f"'{answer}'")
print("chances:", chances)
```

```python
#the main game

while chances>0:
        
    user_input= input("enter a valid letter").lower().strip()
    
    # check validation
    if len(user_input)!=1 or not user_input.isalpha() or user_input.isspace():
        print("enter a single valid letter")
        continue
    

    
    # check correctness
    if user_input in word:
        not_repeated = False
        for i in range(word_length):
            if word[i]==user_input and display[i]=="-":
                display[i]= user_input
                not_repeated = True
        if not_repeated:
            print(f"'{user_input}' is a correct letter")
        else: 
            print (f"you've already guessed the letter '{user_input}'")
        
        
    else:
        print(f"'{user_input}' is wrong , try another letter")
        

    answer = "".join(display)
    print (answer)

    if answer==word:
        print("congrats you found the secret fruit")
        break 

    chances -= 1
    print(f"you still have {chances} chances")
    print("################################################################")
```
